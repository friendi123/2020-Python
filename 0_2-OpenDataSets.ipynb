```python
list1 = [2, 2, 2]
list2 = [1.5, 1.2, 1]
difference = []
# initialization of result list
'''Use zip(iterator1, iterator2) with the lists as iterator1 and iterator2 to return a zip object. 
Use a for-loop to iterate over the zip object and subtract the lists' elements from each other 
and store the result in a list. source: https://kite.com/python/answers/how-to-subtract-two-lists-in-python'''


zip_object = zip(list1, list2)
for list1_i, list2_i in zip_object:
    difference.append(list1_i-list2_i)
    
print(difference)
```

```python
sum_list = []
# initialize result list

for (item1, item2) in zip(list1, list2):
    sum_list.append(item1+item2)

print(sum_list)
```

```python
# Random variables using the scipy package
# scipy has a lot more than just probability distributions; there are sub-libraries for topics such as linear algebra, optimization, etc.
# For more information about scipy see https://docs.scipy.org/doc/scipy/reference/.

from scipy import stats
# plot normal pdf
x = np.linspace(-4,4,100)
y = stats.norm().pdf(x)

import matplotlib.pyplot as plt

plt.plot(x,y)
plt.show()

# plot normal cdf
# (of course, you can also use this for Z tests, etc.)
x = np.linspace(-4,4,100)
y = stats.norm().cdf(x)

plt.plot(x,y)
plt.show()
```

```python
# generate random variables from a distribution
# (note: you can also do this in numpy using np.random)
x = stats.norm(10,4).rvs(3)
y = stats.expon(4).rvs(4)
z = stats.poisson(1).rvs(5)

print(x.round(2))
print(y.round(2))
print(z)
```

```python
# The pandas package provides its special objects and functions for easily reading in and manipulating data.
# For more information see http://pandas.pydata.org/pandas-docs/stable/.

import pandas as pd

x = pd.Series(stats.norm().rvs(10))
print(x.head())

# some of the many available summary statistics
[x.min(), x.quantile(0.25), x.median(), x.quantile(0.75), x.max(), x.mean(), x.var(), x.autocorr()]

# filter
x.where(x > 0)

# filter and remove missing values
x.where(x > 0).dropna()

# As with numpy arrays, operations are elementwise.
x + x

x ** 2

# Concatenate using append.
x2 = x.append(x)
print(x2.head())

# create using a dictionary
data1 = {'x': stats.uniform().rvs(10), 'y': stats.norm().rvs(10)}
df_data1 = pd.DataFrame(data1)
print(df_data1.head())

# create using a list
data2 = [stats.uniform().rvs(10),stats.norm().rvs(10)]
df_data2 = pd.DataFrame(data2)
print(df_data2.head())

# transpose
df_data2T = df_data2.T
print(df_data2T.head())
```

```python
# rename columns
df_data2T.columns = ['x','y']
print(df_data2T)

# pick out a column by name
df_data2T.x

# equivalent: df_data2T['x']

# Note that columns in a pandas DataFrame are Series objects.

type(df_data2T.x)
```

```python
# select columns by index
df2 = df_data2T.iloc[:,1]
print(df2)
```

```python
# select rows by index
df2 = df_data2T.iloc[0:5,]
print(df2)
```

```python
# select rows by value
df2 = df_data2T.where(df_data2T['x']>0.5,).dropna()
print(df2)
```
